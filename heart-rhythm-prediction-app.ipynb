```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('/kaggle/input/mitbih-dataset/mitbih_train.csv')
test = pd.read_csv('/kaggle/input/mitbih-dataset/mitbih_test.csv')
```

```python
train.head()
```

```python
x_train = np.array(train)[:,:187]
y_train = np.array(train)[:,187]
```

```python
x_test = np.array(test)[:,:187]
y_test = np.array(test)[:,187]
```

```python
from sklearn.naive_bayes import CategoricalNB
gnb = CategoricalNB()
gnb.fit(x_train,y_train)
```

```python
tahmin = gnb.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test,tahmin)
index = ['No','S','V','F','Q']
columns = ['No','S','V','F','Q']
```

```python
cm_df = pd.DataFrame(cm,columns,index)
```

```python
plt.figure(figsize = (10,6))
sns.heatmap(cm_df,annot = True,fmt = "d",cmap = "YlGnBu")
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy Score : ",accuracy_score(y_test,tahmin))
```
